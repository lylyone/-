```python
/**
    * 
    *
    * @param array
    * @return
    */
   public static int[] bubbleSort(int[] array) {
       if (array.length == 0)
           return array;
       for (int i = 0; i < array.length; i++)
           for (int j = 0; j < array.length - 1 - i; j++)
               if (array[j + 1] < array[j]) {
                   int temp = array[j + 1];
                   array[j + 1] = array[j];
                   array[j] = temp;
               }
       return array;
   }
```

```python
def bubbleSort(A):
    if len(A)==0:
        return A;
    for i in range(len(A)):
        for j in range(len(A)-i-1):
            if A[j+1]<A[j]:
                temp=A[j+1]
                A[j+1]=A[j]
                A[j]=temp
    return A
                
```

```python
/**
    * 
    * @param array
    * @return
    */
   public static int[] insertionSort(int[] array) {
       if (array.length == 0)
           return array;
       int current;
       for (int i = 0; i < array.length - 1; i++) {
           current = array[i + 1];
           int preIndex = i;
           while (preIndex >= 0 && current < array[preIndex]) {
               array[preIndex + 1] = array[preIndex];
               preIndex--;
           }
           array[preIndex + 1] = current;
       }
       return array;
   }
```

# python

```python
def insertionSort(arr):
    if len(arr)==0:
        return arr;
    for i in range(len(arr)-1):
        temp=arr[i+1]
        pre=i
        while pre>=0 and temp<arr[pre]:
            arr[pre+1]=arr[pre]
            pre-=1
        arr[pre+1]=temp
    return arr
            
```

```python
l=[2,3,1,3,5,78,32,8,3223,5,8,231,12,31,1,2]
insertionSort(l)
```

```python
/**
    * 
    *
    * @param array
    * @return
    */
   public static int[] ShellSort(int[] array) {
       int len = array.length;
       int temp, gap = len / 2;
       while (gap > 0) {
           for (int i = gap; i < len; i++) {
               temp = array[i];
               int preIndex = i - gap;
               while (preIndex >= 0 && array[preIndex] > temp) {
                   array[preIndex + gap] = array[preIndex];
                   preIndex -= gap;
               }
               array[preIndex + gap] = temp;
           }
           gap /= 2;
       }
       return array;
   }
```

#  python

```python
def ShellSort(arr):
    gap=int(len(arr)/2)
    while gap>0:
        for i in range(gap,len(arr)):
            temp=arr[i]
            pre=i-gap
            while pre>=0 and arr[pre]>temp:
                arr[pre+gap]=arr[pre]
                pre-=gap
            arr[pre+gap]=temp
        gap=int(gap/2)
    return arr
```

```python
l=[2,3,1,3,5,78,32,8,3223,5,8,231,12,31,1,2]
ShellSort(l)
```
